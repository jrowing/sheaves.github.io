# Homological algebra recap



## Projective resolutions

Let $R$ be a ring, $M$ an $R$-module, 

A **projective resolution** of $M$ is an exact sequence of $R$-modules

$$ \cdots \to F_2 \xrightarrow{\partial_2} F_1 \xrightarrow{\partial_1} F_0 \xrightarrow{\varepsilon} M \to 0$$

such that each $F_i$ is a projective $R$-module.

Also written:

$$ F_{\bullet} \xrightarrow{\varepsilon} M $$

- Free modules are projective, and we can always get a free, hence projective, resolution
- $\varepsilon$ is an **augmentation** of $F_\bullet$

## Tensors and Tor

Let $N$ be another $R$-module. 

Tensoring $N$ with $F_\bullet$, we get another chain complex

$$ F_{\bullet} \otimes_R N =  \cdots \to F_2 \otimes N \to F_1 \otimes N \to F_0 \otimes N \to 0.$$

Taking the homology of this complex yields:

$$ \text{Tor}^R_n(M,N) := H_n(F_{\bullet}\otimes_R N)$$




- Tensoring is right exact in general
- If $N$ is **flat** then tensoring with $N$ is an exact functor, so all the homology groups are trivial
- Note that we only tensor with $F_\bullet$, not $F_\bullet \to M$
- $\text{Tor}(-,N)$ is the derived functor of $-\otimes N$
- Tor is independent of the resolution of $M$
- We can also define Tor using resolutions of $N$; the results are isomorphic

# Group rings and homology



## $\mathbb{Z}G$-modules

Let $G$ be a (finite) group. Want to study the **integral group ring**

$$R = \mathbb{Z}G$$

and its modules.

- $\mathbb{Z}G$ is the free $\mathbb{Z}$-module generated by elements of $G$, with multiplication extended linearly from that of $G$
- This is the ring we will be using from now on
- If $G$ is not finite, we require that elements of $\mathbb{Z}G$ have only finitely many nonzero $a_g$

Some $\mathbb{Z}G$-modules:
- $\mathbb{Z}X$, where $X$ is a $G$-set
- $\mathbb{Z} \cong \mathbb{Z}\{\bullet\}$
- $\mathbb{Z}G$

## Homology of a group

The **group homology** of $G$ is

$$ H_{\bullet}G := \text{Tor}^{\mathbb{Z}G}_{\bullet}(\mathbb{Z},\mathbb{Z})$$

which is the homology of the chain complex

$$ \cdots \to F_2 \otimes_{\mathbb{Z}G} \mathbb{Z} \to F_1 \otimes_{\mathbb{Z}G} \mathbb{Z} \to F_0 \otimes_{\mathbb{Z}G} \mathbb{Z} \to 0$$

where

$$\cdots \to F_2 \to F_1 \to F_0 \to \mathbb{Z} \to 0$$

is a projective resolution of $\mathbb{Z}$.

#### Topology provides a resolution of $\mathbb{Z}$!
